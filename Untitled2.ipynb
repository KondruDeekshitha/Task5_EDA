```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (8,5)

# Load dataset
df = pd.read_csv("tested.csv")  # change filename if different
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
#missing values
df.isnull().sum()
sns.heatmap(df.isnull(), cbar=False)
```

```python
#univariate analysis
sns.countplot(x='Sex', data=df)
plt.show()

df['Age'].hist(bins=30)
plt.show()
```

```python
#bivariate analysis
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.show()
```

```python
#correlation
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
